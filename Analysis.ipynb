```python
sns.heatmap(df.corr(numeric_only=True),annot=True)
plt.rcParams['figure.figsize']=(30,7)
plt.show()
```

```python
continet_wise_population=df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population',ascending=False)
continet_wise_population
```

```python
df2=df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population',ascending=False)
df2.plot()
```
